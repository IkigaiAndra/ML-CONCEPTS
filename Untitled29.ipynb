```python
import numpy as np

# Generate 20,000 data points with 2 features
X = np.random.rand(20000, 2)
```

```python
from sklearn.decomposition import TruncatedSVD

# Apply SVD to reduce dimensionality to 2 components
svd = TruncatedSVD(n_components=2)
X_transformed = svd.fit_transform(X)
```

```python
import matplotlib.pyplot as plt

# Plot the transformed data
plt.scatter(X_transformed[:, 0], X_transformed[:, 1], s=2, alpha=0.5)
plt.xlabel('Singular Value 1')
plt.ylabel('Singular Value 2')
plt.title('SVD of 20,000 Data Points')
plt.show()
```
