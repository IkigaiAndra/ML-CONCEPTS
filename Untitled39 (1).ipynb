```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from pyod.models.deep_svdd import DeepSVDD

# Generate some data with anomalies
X, y = make_blobs(n_samples=300, centers=2, cluster_std=0.8, random_state=42)
X_train = np.concatenate([X[y == 0], X[y == 1][:5]])  # Normal data
X_test = np.concatenate([X[y == 1][5:], X[y == 0][:10]]) # Introduce some anomalies

# Reshape to (n_samples, n_features)
X_train = X_train.reshape(-1, X_train.shape[1])
X_test = X_test.reshape(-1, X_test.shape[1])
```

```python
# Initialize and train the Deep SAD model
# Get the number of features from the training data
n_features = X_train.shape[1]

# Initialize DeepSVDD with n_features
clf = DeepSVDD(contamination=0.1, n_features=n_features)
clf.fit(X_train)

# Get anomaly scores
y_scores = clf.decision_function(X_test)

# Predict anomalies
y_pred = clf.predict(X_test)
```

```python
# Plot the results
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred, cmap='viridis')
plt.title('Deep SAD Anomaly Detection')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Anomaly Score')
plt.show()
```
