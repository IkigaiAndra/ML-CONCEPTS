POSITIVE REINFORCEMENT LEARNING

We'll consider a simple scenario where an agent learns to navigate a 1D grid world. The agent starts at position 0 and aims to reach the goal at position 10. The agent can move either left or right by one step at a time. Positive feedback is provided when the agent moves closer to the goal, encouraging it to continue in that direction.

```python
import numpy as np
import matplotlib.pyplot as plt

# Environment setup
grid_size = 10
goal_position = grid_size

# Agent's policy (random initially)
policy = np.random.rand(grid_size, 2)  # 2 actions: left or right
policy /= np.sum(policy, axis=1, keepdims=True)  # Normalize probabilities

# Learning parameters
learning_rate = 0.1
discount_factor = 0.9

# Data storage
positions = []
rewards = []

# Training loop
for episode in range(100):
    state = 0  # Initial state
    episode_positions = [state]
    episode_rewards = []

    while state != goal_position:
        # Choose action based on policy
        action = np.random.choice([0, 1], p=policy[state])

        # Take action and observe reward
        if action == 0:  # Move left
            next_state = max(0, state - 1)
        else:  # Move right
            next_state = min(grid_size, state + 1)

        reward = 1 if next_state > state else 0  # Positive reward for moving towards goal

        # Update policy using positive feedback
        policy[state, action] += learning_rate * reward * (discount_factor - policy[state, action])
        policy[state] /= np.sum(policy[state])  # Normalize probabilities

        state = next_state
        episode_positions.append(state)
        episode_rewards.append(reward)

    positions.append(episode_positions)
    rewards.append(episode_rewards)

# Plotting the results
plt.figure(figsize=(10, 5))
max_len = max(len(ep_rewards) for ep_rewards in rewards)  # Find the maximum episode length
cumulative_rewards = [np.pad(np.cumsum(ep_rewards), (0, max_len - len(ep_rewards)), 'constant') for ep_rewards in rewards]

# Plot the cumulative rewards
plt.plot(np.array(cumulative_rewards).T)  # Transpose for plotting

plt.xlabel("Time Step")
plt.ylabel("Cumulative Reward")
plt.title("Cumulative Reward over Time")
plt.show()



for i, episode_positions in enumerate(positions):
    plt.plot(episode_positions, label=f"Episode {i + 1}")
plt.xlabel("Time Step")
plt.ylabel("Agent Position")
plt.title("Agent's Trajectory with Positive Feedback")
plt.legend()
plt.show()

max_len = max(len(ep_rewards) for ep_rewards in rewards)  # Find the maximum episode length
cumulative_rewards = [np.pad(np.cumsum(ep_rewards), (0, max_len - len(ep_rewards)), 'constant') for ep_rewards in rewards]

# Plot the cumulative rewards
plt.plot(np.array(cumulative_rewards).T)  # Transpose for plotting

plt.xlabel("Time Step")
plt.ylabel("Cumulative Reward")
plt.title("Cumulative Reward over Time")
plt.show()
```

NEGATIVE REINFORCEMENT LEARNING

We'll consider a similar scenario as before, with an agent navigating a 1D grid world. However, this time, we'll introduce negative feedback for actions that move the agent away from the goal. This will discourage the agent from taking those actions and encourage it to explore alternative strategies.

```python
import numpy as np
import matplotlib.pyplot as plt

# Environment setup
grid_size = 10
goal_position = grid_size

# Agent's policy (random initially)
policy = np.random.rand(grid_size, 2)  # 2 actions: left or right
policy /= np.sum(policy, axis=1, keepdims=True)  # Normalize probabilities

# Learning parameters
learning_rate = 0.1
discount_factor = 0.9

# Data storage
positions = []
rewards = []

# Training loop
for episode in range(100):
    state = 0  # Initial state
    episode_positions = [state]
    episode_rewards = []

    while state != goal_position:
        # Choose action based on policy
        action = np.random.choice([0, 1], p=policy[state])

        # Take action and observe reward
        if action == 0:  # Move left
            next_state = max(0, state - 1)
        else:  # Move right
            next_state = min(grid_size, state + 1)

        # Negative reward for moving away from goal
        reward = -1 if next_state < state else 0

        # Update policy using negative feedback
        policy[state, action] += learning_rate * reward * (discount_factor - policy[state, action])
        policy[state] = np.clip(policy[state], 0, 1)
        policy[state] /= np.sum(policy[state])  # Normalize probabilities

        state = next_state
        episode_positions.append(state)
        episode_rewards.append(reward)

    positions.append(episode_positions)
    rewards.append(episode_rewards)

# Plotting the results
plt.figure(figsize=(10, 5))
max_len = max(len(ep_rewards) for ep_rewards in rewards)  # Find the maximum episode length
cumulative_rewards = [np.pad(np.cumsum(ep_rewards), (0, max_len - len(ep_rewards)), 'constant') for ep_rewards in rewards]

# Plot the cumulative rewards using the padded array
plt.plot(np.array(cumulative_rewards).T)  # Transpose for plotting

for i, episode_positions in enumerate(positions):
    plt.plot(episode_positions, label=f"Episode {i + 1}")
plt.xlabel("Time Step")
plt.ylabel("Agent Position")
plt.title("Agent's Trajectory with Negative Feedback")
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
max_len = max(len(ep_rewards) for ep_rewards in rewards)  # Find the maximum episode length
padded_rewards = [np.pad(ep_rewards, (0, max_len - len(ep_rewards)), 'constant') for ep_rewards in rewards]

plt.plot(np.cumsum(padded_rewards, axis=1).T)  # Transpose for plotting

plt.xlabel("Time Step")
plt.ylabel("Cumulative Reward")
plt.title("Cumulative Reward over Time")
plt.show()
```
