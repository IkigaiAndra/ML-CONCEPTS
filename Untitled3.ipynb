```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some random data
np.random.seed(0)
X = np.random.rand(100, 1)
y = 2 + 3 * X + np.random.rand(100, 1)

# Split data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Create an overfitted model
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly = PolynomialFeatures(degree=20)  # High degree for overfitting
X_train_poly = poly.fit_transform(X_train)
model_overfit = LinearRegression()
model_overfit.fit(X_train_poly, y_train)

# Create an underfitted model
model_underfit = LinearRegression()
model_underfit.fit(X_train, y_train)

# Generate predictions
X_range = np.linspace(0, 1, 100).reshape(-1, 1)
X_range_poly = poly.transform(X_range)
y_pred_overfit = model_overfit.predict(X_range_poly)
y_pred_underfit = model_underfit.predict(X_range)

# Plot the results
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.scatter(X_test, y_test, color='red', label='Testing Data')
plt.plot(X_range, y_pred_overfit, color='green', label='Overfitted Model')
plt.plot(X_range, y_pred_underfit, color='black', label='Underfitted Model')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Overfitted vs. Underfitted Model')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some random data
np.random.seed(0)
X = np.random.rand(100, 1)
y = 2 + 3 * X + np.random.rand(100, 1)

# Split data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Create an overfitted model (5 loops)
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly_overfit = PolynomialFeatures(degree=5)  # 5 loops for overfitting
X_train_poly_overfit = poly_overfit.fit_transform(X_train)
model_overfit = LinearRegression()
model_overfit.fit(X_train_poly_overfit, y_train)

# Create an underfitted model
model_underfit = LinearRegression()
model_underfit.fit(X_train, y_train)

# Generate predictions
X_range = np.linspace(0, 1, 100).reshape(-1, 1)
X_range_poly_overfit = poly_overfit.transform(X_range)
y_pred_overfit = model_overfit.predict(X_range_poly_overfit)
y_pred_underfit = model_underfit.predict(X_range)

# Plot the results
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.scatter(X_test, y_test, color='red', label='Testing Data')
plt.plot(X_range, y_pred_overfit, color='green', label='Overfitted Model (5 loops)')
plt.plot(X_range, y_pred_underfit, color='black', label='Underfitted Model')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Overfitted (5 loops) vs. Underfitted Model')
plt.legend()
plt.show()
```
