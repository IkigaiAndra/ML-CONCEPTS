self training

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.semi_supervised import SelfTrainingClassifier
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# Generate a synthetic dataset
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, random_state=42)

# Split into labeled and unlabeled data
X_labeled, X_unlabeled, y_labeled, y_unlabeled = train_test_split(X, y, test_size=0.9, random_state=42)

# Create a self-training classifier with Logistic Regression as the base classifier
base_classifier = LogisticRegression()
self_training_model = SelfTrainingClassifier(base_classifier)

# Fit the model to labeled and unlabeled data
self_training_model.fit(np.concatenate((X_labeled, X_unlabeled)), np.concatenate((y_labeled, [-1] * len(y_unlabeled)))) # -1 represents unlabeled data

# Make predictions on the entire dataset
y_pred = self_training_model.predict(X)

# Print the accuracy on the labeled data
print("Accuracy on labeled data:", self_training_model.score(X_labeled, y_labeled))

# Plot the decision boundary and data points
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', marker='o', s=20, edgecolor='k')
plt.title("Self-Training with Logistic Regression")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

co training

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
import matplotlib.pyplot as plt

# Generate a synthetic dataset with two views (features)
X, y = make_classification(n_samples=1000, n_features=4, n_informative=2, n_redundant=2, random_state=42)
X1 = X[:, :2]  # View 1 (first two features)
X2 = X[:, 2:]  # View 2 (last two features)

# Split into labeled and unlabeled data
X1_labeled, X1_unlabeled, X2_labeled, X2_unlabeled, y_labeled, y_unlabeled = train_test_split(
    X1, X2, y, test_size=0.9, random_state=42
)

# Create two classifiers for each view
classifier1 = LogisticRegression()
classifier2 = SVC(probability=True)

# Co-training loop
iterations = 5
for i in range(iterations):
    # Train classifiers on labeled data
    classifier1.fit(X1_labeled, y_labeled)
    classifier2.fit(X2_labeled, y_labeled)

    # Predict labels for unlabeled data
    y1_pred = classifier1.predict_proba(X1_unlabeled)
    y2_pred = classifier2.predict_proba(X2_unlabeled)

    # Select most confident predictions from each classifier
    n_add = 50  # Number of samples to add per iteration
    confident_indices1 = np.argsort(y1_pred[:, 1])[-n_add:]
    confident_indices2 = np.argsort(y2_pred[:, 1])[-n_add:]

    # Add confident predictions to labeled data
    X1_labeled = np.concatenate((X1_labeled, X1_unlabeled[confident_indices1]))
    X2_labeled = np.concatenate((X2_labeled, X2_unlabeled[confident_indices2]))
    y_labeled = np.concatenate((y_labeled, np.ones(n_add)))  # Assuming confident predictions are positive class

    # Remove added samples from unlabeled data
    X1_unlabeled = np.delete(X1_unlabeled, confident_indices1, axis=0)
    X2_unlabeled = np.delete(X2_unlabeled, confident_indices2, axis=0)

# Print accuracy on labeled data for both classifiers
print("Classifier 1 Accuracy on labeled data:", classifier1.score(X1_labeled, y_labeled))
print("Classifier 2 Accuracy on labeled data:", classifier2.score(X2_labeled, y_labeled))

# Plot the data (using only view 1 for visualization)
plt.figure(figsize=(8, 6))
plt.scatter(X1[:, 0], X1[:, 1], c=classifier1.predict(X1), cmap='viridis', marker='o', s=20, edgecolor='k')
plt.title("Co-Training with Logistic Regression and SVM")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

tsvm

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Generate a synthetic dataset
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, random_state=42)

# Split into labeled and unlabeled data
X_labeled, X_unlabeled, y_labeled, y_unlabeled = train_test_split(X, y, test_size=0.9, random_state=42)

# Create a TSVM classifier
tsvm = SVC(kernel='rbf', probability=True)  # Using an RBF kernel for non-linearity

# Fit the model to labeled and unlabeled data
# For TSVM, we fit on the entire dataset, with unlabeled data having labels set to -1
tsvm.fit(np.concatenate((X_labeled, X_unlabeled)), np.concatenate((y_labeled, [-1] * len(y_unlabeled))))

# Predict labels for the entire dataset
y_pred = tsvm.predict(X)

# Print the accuracy on the labeled data
print("Accuracy on labeled data:", accuracy_score(y_labeled, tsvm.predict(X_labeled)))

# Plot the decision boundary and data points
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', marker='o', s=20, edgecolor='k')
plt.title("Transductive SVM (TSVM)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

graph based methods

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.semi_supervised import LabelSpreading
import matplotlib.pyplot as plt

# Generate a synthetic dataset
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, random_state=42)

# Split into labeled and unlabeled data
X_labeled, X_unlabeled, y_labeled, y_unlabeled = train_test_split(X, y, test_size=0.9, random_state=42)

# Create a LabelSpreading model (a graph-based method)
label_spreading = LabelSpreading(kernel='rbf', gamma=20, max_iter=10)  # Using an RBF kernel for similarity

# Fit the model to labeled and unlabeled data
label_spreading.fit(np.concatenate((X_labeled, X_unlabeled)), np.concatenate((y_labeled, [-1] * len(y_unlabeled))))

# Predict labels for the entire dataset
y_pred = label_spreading.predict(X)

# Print the accuracy on the labeled data
print("Accuracy on labeled data:", label_spreading.score(X_labeled, y_labeled))

# Plot the decision boundary and data points
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', marker='o', s=20, edgecolor='k')
plt.title("Label Spreading (Graph-Based Method)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
