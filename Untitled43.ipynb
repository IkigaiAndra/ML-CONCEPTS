```python
import numpy as np
import random
import matplotlib.pyplot as plt

# Gridworld size (3D grid)
grid_size = 10  # Use a smaller size for visualization purposes (you can extend this for larger grids)
goal_position = (grid_size - 1, grid_size - 1, grid_size - 1)  # Goal is at the opposite corner

# Q-learning parameters
epsilon = 1.0  # Exploration rate
epsilon_min = 0.01
epsilon_decay = 0.995
gamma = 0.9  # Discount factor
learning_rate = 0.1  # Learning rate
num_episodes = 100  # Number of episodes
max_steps = 100  # Maximum steps per episode

# Q-table initialization
Q = np.zeros((grid_size, grid_size, grid_size, 6))  # 6 actions: up, down, left, right, forward, backward

# Environment dynamics (actions)
# The six possible actions are:
# 0: Move up (increase z-coordinate), 1: Move down (decrease z-coordinate)
# 2: Move left (decrease x-coordinate), 3: Move right (increase x-coordinate)
# 4: Move forward (increase y-coordinate), 5: Move backward (decrease y-coordinate)
action_map = {
    0: (0, 0, 1),  # Move up
    1: (0, 0, -1),  # Move down
    2: (-1, 0, 0),  # Move left
    3: (1, 0, 0),   # Move right
    4: (0, -1, 0),  # Move forward
    5: (0, 1, 0)    # Move backward
}

# Reward function: Positive feedback as the agent moves closer to the goal
def step(state, action):
    x, y, z = state
    dx, dy, dz = action_map[action]

    # Calculate the next state
    next_state = (min(max(0, x + dx), grid_size - 1),
                  min(max(0, y + dy), grid_size - 1),
                  min(max(0, z + dz), grid_size - 1))

    # Reward: Positive feedback as the agent moves closer to the goal
    goal_distance = np.linalg.norm(np.array(next_state) - np.array(goal_position))
    reward = max(0, 10 - goal_distance)  # Reward is inversely proportional to the distance to the goal

    return next_state, reward

# Epsilon-greedy policy to select an action
def epsilon_greedy(state, epsilon):
    if np.random.rand() <= epsilon:
        return random.choice([0, 1, 2, 3, 4, 5])  # Random action (exploration)
    else:
        return np.argmax(Q[state[0], state[1], state[2]])  # Greedy action (exploitation)

# Train the agent using Q-learning with positive feedback
def train_qlearning():
    global epsilon
    rewards_per_episode = []

    for episode in range(num_episodes):
        state = (0, 0, 0)  # Start at the origin of the grid
        total_reward = 0
        done = False

        for step_count in range(max_steps):
            action = epsilon_greedy(state, epsilon)
            next_state, reward = step(state, action)
            total_reward += reward

            # Q-value update
            best_next_action = np.argmax(Q[next_state[0], next_state[1], next_state[2]])  # Best action at next state
            Q[state[0], state[1], state[2], action] = Q[state[0], state[1], state[2], action] + learning_rate * (reward + gamma * Q[next_state[0], next_state[1], next_state[2], best_next_action] - Q[state[0], state[1], state[2], action])

            state = next_state  # Update the state

            if state == goal_position:
                done = True
                break

        rewards_per_episode.append(total_reward)

        # Decay epsilon to reduce exploration over time
        if epsilon > epsilon_min:
            epsilon *= epsilon_decay

    return rewards_per_episode

# Train the agent and collect rewards per episode
rewards = train_qlearning()

# Plotting the reward over episodes
plt.plot(rewards)
plt.title("Reward Over Episodes with Positive Feedback (Q-Learning in 3D Gridworld)")
plt.xlabel("Episodes")
plt.ylabel("Total Reward")
plt.show()
```
