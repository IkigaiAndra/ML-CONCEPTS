```python
import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate 20,000 data points
n_samples = 20000
X, y = make_blobs(n_samples=n_samples, centers=5, random_state=42)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(X)

# Create a Pandas DataFrame for easier plotting
df = pd.DataFrame(X, columns=['x', 'y'])
df['cluster'] = clusters

# Plot the clusters in detail
plt.figure(figsize=(10, 8))
for cluster in df['cluster'].unique():
    plt.scatter(df[df['cluster'] == cluster]['x'],
                df[df['cluster'] == cluster]['y'],
                label=f'Cluster {cluster}', alpha=0.7)

plt.title('DBSCAN Clustering with 20,000 Data Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```
