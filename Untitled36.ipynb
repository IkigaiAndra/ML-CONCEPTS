```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Generate synthetic text data
n_samples = 1000
n_features = 100  # Number of words in the vocabulary
n_classes = 2  # Binary classification

# Create random text data (replace with your actual text data)
rng = np.random.RandomState(42)
X = rng.randint(0, n_features, size=(n_samples, 5))  # Each sample has 5 random words
y = rng.randint(0, n_classes, size=n_samples)

# Convert integer data to text data
vocabulary = [f"word_{i}" for i in range(n_features)]
text_data = [
    " ".join([vocabulary[word_index] for word_index in sample]) for sample in X
]

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    text_data, y, test_size=0.2, random_state=42
)

# Create a TF-IDF vectorizer to convert text to numerical features
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Train a Logistic Regression classifier
classifier = LogisticRegression()
classifier.fit(X_train_tfidf, y_train)

# Predict labels for the test set
y_pred = classifier.predict(X_test_tfidf)

# Evaluate the classifier
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print(classification_report(y_test, y_pred))

# Plot confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.imshow(cm, interpolation="nearest", cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.colorbar()
plt.xticks([0, 1], ["Class 0", "Class 1"])
plt.yticks([0, 1], ["Class 0", "Class 1"])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```
