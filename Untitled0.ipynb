```python
//importing libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Generate random data for demonstration
np.random.seed(0)  # Set seed for reproducibility
X = np.random.rand(100, 1) * 10  # Independent variable
y = 2 * X + 3 + np.random.randn(100, 1) * 2  # Dependent variable with noise
```

```python
model = LinearRegression()  # Create a LinearRegression object
model.fit(X, y)  # Fit the model to the data
```

```python
y_pred = model.predict(X)  # Predict values using the trained model
```

```python
plt.scatter(X, y, label='Data')  # Scatter plot of the original data
plt.plot(X, y_pred, color='red', label='Regression Line')  # Plot the regression line
plt.xlabel('X')  # Label for the x-axis
plt.ylabel('y')  # Label for the y-axis
plt.title('Linear Regression')  # Title of the plot
plt.legend()  # Show legend
plt.show()  # Display the plot
```
