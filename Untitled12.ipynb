```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate random data from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)
```

```python
# Plot the histogram of the data
sns.histplot(data, kde=True, stat="density", color='skyblue')

# Plot the Gaussian function
x = np.linspace(-3, 3, 100)  # Create x-axis values
gaussian_function = (1 / (np.sqrt(2 * np.pi))) * np.exp(-0.5 * x**2)  # Gaussian formula
plt.plot(x, gaussian_function, color='red', label='Gaussian Function')

# Customize the plot
plt.title('Gaussian Distribution')
plt.xlabel('Data Values')
plt.ylabel('Density')
plt.legend()
plt.show()
```
