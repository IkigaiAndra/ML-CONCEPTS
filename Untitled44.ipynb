Q LEARNING

```python
import numpy as np
import matplotlib.pyplot as plt

# Environment definition: 5x5 grid
grid_size = 5
n_actions = 4  # Up, Down, Left, Right
goal_state = (4, 4)  # Goal position in the grid
start_state = (0, 0)  # Start position in the grid

# Define actions
actions = ["up", "down", "left", "right"]

# Q-table (State-action value table)
q_table = np.zeros((grid_size, grid_size, n_actions))

# Hyperparameters
alpha = 0.1  # Learning rate
gamma = 0.9  # Discount factor
epsilon = 0.1  # Exploration rate
episodes = 1000  # Number of episodes
max_steps = 100  # Max steps per episode

# Reward and penalty structure
def get_reward(state):
    if state == goal_state:
        return 100  # Reward for reaching the goal
    else:
        return -1  # Penalty for each step taken

# Function to take a step and return the next state
def take_action(state, action):
    if action == 0:  # Move up
        next_state = (max(0, state[0] - 1), state[1])
    elif action == 1:  # Move down
        next_state = (min(grid_size - 1, state[0] + 1), state[1])
    elif action == 2:  # Move left
        next_state = (state[0], max(0, state[1] - 1))
    elif action == 3:  # Move right
        next_state = (state[0], min(grid_size - 1, state[1] + 1))
    return next_state

# Q-learning algorithm
def q_learning():
    rewards_per_episode = []  # List to store rewards per episode

    for episode in range(episodes):
        state = start_state  # Start at the initial state
        total_reward = 0

        for _ in range(max_steps):
            # Choose action (epsilon-greedy)
            if np.random.rand() < epsilon:
                action = np.random.choice(n_actions)  # Explore
            else:
                action = np.argmax(q_table[state[0], state[1]])  # Exploit

            next_state = take_action(state, action)
            reward = get_reward(next_state)
            total_reward += reward

            # Q-value update using the Q-learning formula
            old_q_value = q_table[state[0], state[1], action]
            future_q_value = np.max(q_table[next_state[0], next_state[1]])

            # Q-learning update rule
            q_table[state[0], state[1], action] = old_q_value + alpha * (reward + gamma * future_q_value - old_q_value)

            # Move to the next state
            state = next_state

            if state == goal_state:
                break

        rewards_per_episode.append(total_reward)

    return rewards_per_episode

# Run Q-learning algorithm
rewards = q_learning()

# Plotting the cumulative rewards over episodes
plt.plot(np.cumsum(rewards))
plt.xlabel('Episodes')
plt.ylabel('Cumulative Reward')
plt.title('Q-Learning Performance')
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Environment definition: 5x5 grid
grid_size = 5
n_actions = 4  # Up, Down, Left, Right
goal_state = (4, 4)  # Goal position in the grid
start_state = (0, 0)  # Start position in the grid

# Define actions
actions = ["up", "down", "left", "right"]

# Q-table (State-action value table)
q_table = np.zeros((grid_size, grid_size, n_actions))

# Hyperparameters
alpha = 0.1  # Learning rate
gamma = 0.9  # Discount factor
epsilon = 0.1  # Exploration rate
episodes = 1000  # Number of episodes
max_steps = 100  # Max steps per episode

# Reward and penalty structure
def get_reward(state):
    if state == goal_state:
        return 100  # Reward for reaching the goal
    else:
        return -1  # Penalty for each step taken

# Function to take a step and return the next state
def take_action(state, action):
    if action == 0:  # Move up
        next_state = (max(0, state[0] - 1), state[1])
    elif action == 1:  # Move down
        next_state = (min(grid_size - 1, state[0] + 1), state[1])
    elif action == 2:  # Move left
        next_state = (state[0], max(0, state[1] - 1))
    elif action == 3:  # Move right
        next_state = (state[0], min(grid_size - 1, state[1] + 1))
    return next_state

# SARSA algorithm
def sarsa():
    rewards_per_episode = []  # List to store rewards per episode

    for episode in range(episodes):
        state = start_state  # Start at the initial state
        total_reward = 0

        # Choose initial action using epsilon-greedy policy
        if np.random.rand() < epsilon:
            action = np.random.choice(n_actions)  # Explore
        else:
            action = np.argmax(q_table[state[0], state[1]])  # Exploit

        for _ in range(max_steps):
            next_state = take_action(state, action)
            reward = get_reward(next_state)
            total_reward += reward

            # Choose next action using epsilon-greedy policy
            if np.random.rand() < epsilon:
                next_action = np.random.choice(n_actions)  # Explore
            else:
                next_action = np.argmax(q_table[next_state[0], next_state[1]])  # Exploit

            # SARSA Q-value update
            old_q_value = q_table[state[0], state[1], action]
            future_q_value = q_table[next_state[0], next_state[1], next_action]

            # SARSA update rule
            q_table[state[0], state[1], action] = old_q_value + alpha * (reward + gamma * future_q_value - old_q_value)

            # Move to the next state and action
            state, action = next_state, next_action

            if state == goal_state:
                break

        rewards_per_episode.append(total_reward)

    return rewards_per_episode

# Run SARSA algorithm
rewards = sarsa()

# Plotting the cumulative rewards over episodes
plt.plot(np.cumsum(rewards))
plt.xlabel('Episodes')
plt.ylabel('Cumulative Reward')
plt.title('SARSA Learning Performance')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Environment definition: 5x5 grid
grid_size = 5
n_actions = 4  # Up, Down, Left, Right
goal_state = (4, 4)  # Goal position in the grid
start_state = (0, 0)  # Start position in the grid

# Define actions
actions = ["up", "down", "left", "right"]

# Hyperparameters
alpha = 0.01  # Learning rate (for policy gradient)
gamma = 0.9  # Discount factor
episodes = 1000  # Number of episodes
max_steps = 100  # Max steps per episode

# Reward and penalty structure
def get_reward(state):
    if state == goal_state:
        return 100  # Reward for reaching the goal
    else:
        return -1  # Penalty for each step taken

# Function to take a step and return the next state
def take_action(state, action):
    if action == 0:  # Move up
        next_state = (max(0, state[0] - 1), state[1])
    elif action == 1:  # Move down
        next_state = (min(grid_size - 1, state[0] + 1), state[1])
    elif action == 2:  # Move left
        next_state = (state[0], max(0, state[1] - 1))
    elif action == 3:  # Move right
        next_state = (state[0], min(grid_size - 1, state[1] + 1))
    return next_state

# Initialize policy (softmax over actions)
theta = np.random.randn(n_actions)  # Parameters for each action (initial random values)

# Softmax policy function
def softmax_policy(state):

   theta_clipped = np.clip(theta, -10, 10)  # Clip values between -10 and 10
   exp_theta = np.exp(theta_clipped - np.max(theta_clipped))
   return exp_theta / np.sum(exp_theta)


   exp_theta = np.exp(theta - np.max(theta))  # Subtract max to prevent overflow
   return exp_theta / np.sum(exp_theta)

# REINFORCE Algorithm
def reinforce():
    global theta  # Declare theta as global to modify it within the function
    rewards_per_episode = []  # List to store rewards per episode
    for episode in range(episodes):
        state = start_state
        episode_rewards = []  # List of rewards for this episode
        episode_actions = []  # List of actions taken in this episode
        episode_states = []  # List of states visited in this episode

        # Collect trajectory (state, action, reward)
        for _ in range(max_steps):
            # Choose action based on softmax policy
            action_probs = softmax_policy(state)
            action = np.random.choice(n_actions, p=action_probs)

            next_state = take_action(state, action)
            reward = get_reward(next_state)

            episode_states.append(state)
            episode_actions.append(action)
            episode_rewards.append(reward)

            state = next_state

            if state == goal_state:
                break

        # Compute the returns (discounted future rewards)
        G = 0
        returns = []
        for r in reversed(episode_rewards):
            G = r + gamma * G
            returns.insert(0, G)

        # Update policy using REINFORCE
        for state, action, G in zip(episode_states, episode_actions, returns):
            action_probs = softmax_policy(state)
            grad_log_pi = np.zeros(n_actions)
            grad_log_pi[action] = 1  # Gradient of the log probability of the taken action
            grad_log_pi /= action_probs[action]

            # Update policy parameters
            theta += alpha * G * grad_log_pi  # Policy gradient update rule

        rewards_per_episode.append(np.sum(episode_rewards))

    return rewards_per_episode

# Run REINFORCE algorithm
rewards = reinforce()

# Plotting the cumulative rewards over episodes
plt.plot(np.cumsum(rewards))
plt.xlabel('Episodes')
plt.ylabel('Cumulative Reward')
plt.title('REINFORCE Learning Performance')
plt.grid(True)
plt.show
```

A3C LIKE

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# Environment definition: 5x5 grid
grid_size = 5
n_actions = 4  # Up, Down, Left, Right
goal_state = (4, 4)  # Goal position in the grid
start_state = (0, 0)  # Start position in the grid

# Define actions
actions = ["up", "down", "left", "right"]

# Hyperparameters
alpha = 0.01  # Learning rate (for policy and value function)
gamma = 0.9  # Discount factor
episodes = 1000  # Number of episodes
max_steps = 100  # Max steps per episode

# Reward and penalty structure
def get_reward(state):
    if state == goal_state:
        return 100  # Reward for reaching the goal
    else:
        return -1  # Penalty for each step taken

# Function to take a step and return the next state
def take_action(state, action):
    if action == 0:  # Move up
        next_state = (max(0, state[0] - 1), state[1])
    elif action == 1:  # Move down
        next_state = (min(grid_size - 1, state[0] + 1), state[1])
    elif action == 2:  # Move left
        next_state = (state[0], max(0, state[1] - 1))
    elif action == 3:  # Move right
        next_state = (state[0], min(grid_size - 1, state[1] + 1))
    return next_state

# Define the Actor-Critic Network
class A3CNet(nn.Module):
    def __init__(self):
        super(A3CNet, self).__init__()
        self.fc = nn.Linear(2, 128)  # Fully connected layer for state representation (x, y)
        self.policy = nn.Linear(128, n_actions)  # Policy network (actor)
        self.value = nn.Linear(128, 1)  # Value network (critic)

    def forward(self, state):
        x = torch.relu(self.fc(state))
        policy_dist = torch.softmax(self.policy(x), dim=-1)
        value = self.value(x)
        return policy_dist, value

# Initialize the network and optimizer
model = A3CNet()
optimizer = optim.Adam(model.parameters(), lr=alpha)

# Function to compute the loss and perform the update
def compute_loss_and_update(state, action, reward, next_state, done, gamma):
    # Get the current policy and value from the model
    state_tensor = torch.tensor(state, dtype=torch.float32).unsqueeze(0)
    policy, value = model(state_tensor)

    # Get the log probability of the action taken
    action_prob = policy[0, action]
    log_prob = torch.log(action_prob)

    # Compute the target value using the reward and next state value (bootstrapping)
    next_state_tensor = torch.tensor(next_state, dtype=torch.float32).unsqueeze(0)
    _, next_value = model(next_state_tensor)

    target_value = reward + gamma * next_value * (1 - int(done))

    # Compute the advantage (TD error)
    advantage = target_value - value

    # Actor (Policy) loss: Negative log probability times advantage
    actor_loss = -log_prob * advantage

    # Critic (Value) loss: Mean squared error of the value estimate
    critic_loss = advantage.pow(2)

    # Total loss
    total_loss = actor_loss + critic_loss

    # Backpropagate and update the weights
    optimizer.zero_grad()
    total_loss.backward()
    optimizer.step()

# A3C-like Training Loop
def a3c():
    rewards_per_episode = []  # List to store rewards per episode
    for episode in range(episodes):
        state = start_state
        total_reward = 0
        done = False

        for _ in range(max_steps):
            # Choose action from policy
            state_tensor = torch.tensor(state, dtype=torch.float32).unsqueeze(0)
            policy, _ = model(state_tensor)
            action = np.random.choice(n_actions, p=policy.detach().numpy().flatten())

            # Take the action
            next_state = take_action(state, action)
            reward = get_reward(next_state)
            total_reward += reward

            # Compute the loss and update the model
            done = (next_state == goal_state)
            compute_loss_and_update(state, action, reward, next_state, done, gamma)

            # Move to the next state
            state = next_state

            if done:
                break

        rewards_per_episode.append(total_reward)

    return rewards_per_episode

# Run A3C-like algorithm
rewards = a3c()

# Plotting the cumulative rewards over episodes
plt.plot(np.cumsum(rewards))
plt.xlabel('Episodes')
plt.ylabel('Cumulative Reward')
plt.title('A3C-like Learning Performance')
plt.grid(True)
plt.show()
```

DDPG

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Parameters
n_actions = 1  # Continuous action space (e.g., one-dimensional control)
n_states = 10  # Number of states (discretized state space)
episodes = 100  # Number of episodes for training
gamma = 0.99  # Discount factor for future rewards
alpha_actor = 0.001  # Learning rate for the actor
alpha_critic = 0.005  # Learning rate for the critic
tau = 0.005  # Target network update rate

# Neural Network for Actor (deterministic action output)
class Actor(tf.keras.Model):
    def __init__(self):
        super(Actor, self).__init__()
        self.dense1 = tf.keras.layers.Dense(128, activation='relu')
        self.dense2 = tf.keras.layers.Dense(64, activation='relu')
        self.output_layer = tf.keras.layers.Dense(n_actions, activation='tanh')  # Action output (normalized to [-1, 1])

    def call(self, state):
        x = self.dense1(state)
        x = self.dense2(x)
        return self.output_layer(x)

# Neural Network for Critic (state-action value)
class Critic(tf.keras.Model):
    def __init__(self):
        super(Critic, self).__init__()
        self.dense1 = tf.keras.layers.Dense(128, activation='relu')
        self.dense2 = tf.keras.layers.Dense(64, activation='relu')
        self.output_layer = tf.keras.layers.Dense(1)  # Value of state-action pair

    def call(self, state, action):
        x = tf.concat([state, action], axis=-1)  # Combine state and action
        x = self.dense1(x)
        x = self.dense2(x)
        return self.output_layer(x)

# Target networks (soft updates from main networks)
def update_target_network(main_network, target_network, tau):
    for target_param, main_param in zip(target_network.variables, main_network.variables):
        target_param.assign(tau * main_param + (1 - tau) * target_param)

# Generate random reward and state transitions
def get_reward(state, action):
    return np.random.randn()  # Random reward from a normal distribution

# Choose an action based on the policy (Actor)
def choose_action(actor, state):
    state = np.expand_dims(state, axis=0)  # Add batch dimension
    action = actor(state)  # Get action from Actor
    return action.numpy()[0]

# Experience replay memory (FIFO buffer)
class ReplayBuffer:
    def __init__(self, max_size=1000000):
        self.buffer = []
        self.max_size = max_size

    def add(self, experience):
        if len(self.buffer) >= self.max_size:
            self.buffer.pop(0)
        self.buffer.append(experience)

    def sample(self, batch_size):
      buffer_array = np.array(self.buffer, dtype=object)
      indices = np.random.choice(len(buffer_array), batch_size, replace=False)
      return buffer_array[indices]


# Initialize networks and optimizers
actor = Actor()
critic = Critic()
actor_target = Actor()
critic_target = Critic()

actor_optimizer = tf.keras.optimizers.Adam(learning_rate=alpha_actor)
critic_optimizer = tf.keras.optimizers.Adam(learning_rate=alpha_critic)

# Replay buffer
replay_buffer = ReplayBuffer(max_size=50000)

reward_history = []

# Training loop for DDPG
for episode in range(episodes):
    state = np.random.rand(n_states)  # Random initial state
    total_reward = 0
    done = False

    for step in range(100):  # Limit steps per episode
        action = choose_action(actor, state)
        reward = get_reward(state, action)
        total_reward += reward

        # Simulate transition to the next state
        next_state = np.random.rand(n_states)  # Random next state

        # Add experience to the replay buffer
        replay_buffer.add((state, action, reward, next_state))

        # Sample a batch from the replay buffer
        if len(replay_buffer.buffer) > 64:
            batch = replay_buffer.sample(64)
            states, actions, rewards, next_states = zip(*batch)
            states = np.array(states)
            actions = np.array(actions).reshape(-1, 1)
            rewards = np.array(rewards).reshape(-1, 1)
            next_states = np.array(next_states)

            # Compute target Q values
            target_actions = actor_target(next_states)
            target_q_values = critic_target(next_states, target_actions)
            target_q_values = rewards + gamma * target_q_values

            # Update Critic network (MSE loss)
            with tf.GradientTape() as tape:
                q_values = critic(states, actions)
                critic_loss = tf.reduce_mean(tf.square(target_q_values - q_values))
            critic_grads = tape.gradient(critic_loss, critic.trainable_variables)
            critic_optimizer.apply_gradients(zip(critic_grads, critic.trainable_variables))

            # Update Actor network (policy gradient)
            with tf.GradientTape() as tape:
                actions = actor(states)
                critic_value = critic(states, actions)
                actor_loss = -tf.reduce_mean(critic_value)  # Maximize Q value for chosen actions
            actor_grads = tape.gradient(actor_loss, actor.trainable_variables)
            actor_optimizer.apply_gradients(zip(actor_grads, actor.trainable_variables))

            # Update target networks with soft updates
            update_target_network(actor, actor_target, tau)
            update_target_network(critic, critic_target, tau)

        # Move to the next state
        state = next_state

    reward_history.append(total_reward)

# Plotting the rewards history over episodes
plt.figure(figsize=(10, 6))
plt.plot(reward_history)
plt.xlabel("Episodes")
plt.ylabel("Total Reward")
plt.title("DDPG Reward History over 100 Episodes")
plt.show()
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Parameters
n_actions = 4  # Number of possible actions
n_states = 10  # Number of possible states
episodes = 10  # Number of episodes for training
gamma = 0.99  # Discount factor for future rewards
alpha_actor = 0.001  # Learning rate for the actor
alpha_critic = 0.005  # Learning rate for the critic

# Neural Network for Actor
class Actor(tf.keras.Model):
    def __init__(self):
        super(Actor, self).__init__()
        self.dense1 = tf.keras.layers.Dense(128, activation='relu')
        self.dense2 = tf.keras.layers.Dense(64, activation='relu')
        self.output_layer = tf.keras.layers.Dense(n_actions, activation='softmax')  # Probabilities for each action

    def call(self, state):
        x = self.dense1(state)
        x = self.dense2(x)
        return self.output_layer(x)

# Neural Network for Critic
class Critic(tf.keras.Model):
    def __init__(self):
        super(Critic, self).__init__()
        self.dense1 = tf.keras.layers.Dense(128, activation='relu')
        self.dense2 = tf.keras.layers.Dense(64, activation='relu')
        self.output_layer = tf.keras.layers.Dense(1)  # Value of the state

    def call(self, state):
        x = self.dense1(state)
        x = self.dense2(x)
        return self.output_layer(x)

# Generate random reward and state transitions
def get_reward(state, action):
    return np.random.randn()  # Random reward from a normal distribution

# Choose an action based on the policy (Actor)
def choose_action(actor, state):
    state = np.expand_dims(state, axis=0)  # Add batch dimension
    action_probs = actor(state)  # Get action probabilities
    action = np.random.choice(n_actions, p=action_probs[0].numpy())  # Sample action
    return action, action_probs

# Training loop for A2C
actor = Actor()
critic = Critic()
actor_optimizer = tf.keras.optimizers.Adam(learning_rate=alpha_actor)
critic_optimizer = tf.keras.optimizers.Adam(learning_rate=alpha_critic)

reward_history = []

for episode in range(episodes):
    state = np.random.rand(n_states)  # Random initial state
    total_reward = 0
    done = False

    # For each episode, run for a set number of steps
    for step in range(100):  # Limit steps per episode
        action, action_probs = choose_action(actor, state)
        reward = get_reward(state, action)
        total_reward += reward

        # Calculate value of current state and next state
        state_value = critic(state.reshape(1, -1))  # Critic evaluates the current state
        next_state = np.random.rand(n_states)  # Random next state
        next_state_value = critic(next_state.reshape(1, -1))  # Critic evaluates the next state

        # Calculate advantage
        advantage = reward + gamma * next_state_value - state_value

        # Update Critic: Minimize the loss (TD error)
        with tf.GradientTape() as tape:
            value = critic(state.reshape(1, -1))
            critic_loss = tf.reduce_mean((reward + gamma * next_state_value - value) ** 2)
        critic_grads = tape.gradient(critic_loss, critic.trainable_variables)
        critic_optimizer.apply_gradients(zip(critic_grads, critic.trainable_variables))

        # Update Actor: Maximize the log probability of the action, weighted by the advantage
        with tf.GradientTape() as tape:
            action_probs = actor(state.reshape(1, -1))
            action_prob = action_probs[0, action]
            actor_loss = -tf.math.log(action_prob) * advantage
        actor_grads = tape.gradient(actor_loss, actor.trainable_variables)
        actor_optimizer.apply_gradients(zip(actor_grads, actor.trainable_variables))

        # Move to the next state
        state = next_state

    reward_history.append(total_reward)

# Plotting the rewards history over episodes
plt.figure(figsize=(10, 6))
plt.plot(reward_history)
plt.xlabel("Episodes")
plt.ylabel("Total Reward")
plt.title("A2C Reward History over 10 Episodes")
plt.show()
```
