```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
# Generate 20,000 data points with 5 clusters
n_samples = 20000
n_features = 2  # Number of features (for visualization)
n_clusters = 5  # Number of clusters

X, y = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters, random_state=42)
# Set random_state for reproducibility

# Create a pandas DataFrame for easier handling
data = pd.DataFrame(X, columns=['feature1', 'feature2'])
data['cluster'] = y  # Add cluster labels to the DataFrame
```

```python
# Select features for clustering
X_cluster = data[['feature1', 'feature2']]

# Create a KMeans object and fit it to the data
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(X_cluster)

# Get cluster assignments for each data point
labels = kmeans.labels_
```

```python
# Plot the data points colored by their assigned clusters
plt.scatter(data['feature1'], data['feature2'], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=200, c='red') # Plot cluster centers
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering on Generated Data')
plt.show()
```
