POLICY GRADIENT

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

# Step 1: Generate random self-generated data (simulating rewards and actions)
np.random.seed(0)  # for reproducibility
num_data = 10000  # Number of data points
states = np.random.randn(num_data, 1)  # Random states
actions = np.random.randint(0, 2, num_data)  # Random actions (0 or 1)
rewards = np.random.randn(num_data)  # Random rewards (continuous)

# Step 2: Define the policy network
class PolicyNetwork(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(PolicyNetwork, self).__init__()
        self.fc = nn.Linear(input_dim, 64)  # Fully connected layer
        self.fc2 = nn.Linear(64, output_dim)  # Output layer
        self.softmax = nn.Softmax(dim=-1)  # Softmax to get probabilities of actions

    def forward(self, x):
        x = torch.relu(self.fc(x))
        return self.softmax(self.fc2(x))

# Step 3: Define Policy Gradient function
def policy_gradient(env_data, model, optimizer, gamma=0.99):
    states, actions, rewards = env_data
    discounted_rewards = []
    R = 0
    for r in reversed(rewards):
        R = r + gamma * R
        discounted_rewards.insert(0, R)

    discounted_rewards = torch.tensor(discounted_rewards, dtype=torch.float32)
    states = torch.tensor(states, dtype=torch.float32)
    actions = torch.tensor(actions, dtype=torch.int64)

    # Normalize the rewards for stability
    discounted_rewards = (discounted_rewards - discounted_rewards.mean()) / (discounted_rewards.std() + 1e-7)

    optimizer.zero_grad()

    # Forward pass through the policy network
    probs = model(states)

    # Select the probabilities corresponding to the actions
    action_probs = probs.gather(1, actions.view(-1, 1))

    # Compute the loss
    loss = -torch.log(action_probs) * discounted_rewards

    # Backpropagate the loss
    loss.mean().backward()
    optimizer.step()

# Step 4: Initialize model, optimizer, and training loop
input_dim = 1  # Input dimension (state)
output_dim = 2  # Output dimension (action space: 0 or 1)
model = PolicyNetwork(input_dim, output_dim)
optimizer = optim.Adam(model.parameters(), lr=1e-3)

# Training loop
num_episodes = 1000
episode_rewards = []

for episode in range(num_episodes):
    # Sample random data for each episode
    indices = np.random.choice(num_data, size=100, replace=False)
    sampled_states = states[indices]
    sampled_actions = actions[indices]
    sampled_rewards = rewards[indices]

    # Train the policy gradient model
    policy_gradient((sampled_states, sampled_actions, sampled_rewards), model, optimizer)

    # Evaluate the model and store rewards
    with torch.no_grad():
        probs = model(torch.tensor(sampled_states, dtype=torch.float32))
        action_probs = probs.gather(1, torch.tensor(sampled_actions, dtype=torch.int64).view(-1, 1))
        reward = (action_probs * torch.tensor(sampled_rewards, dtype=torch.float32)).sum().item()

    episode_rewards.append(reward)

# Step 5: Plot the results (reward per episode)
plt.plot(episode_rewards)
plt.title("Policy Gradient Learning Progress")
plt.xlabel("Episode")
plt.ylabel("Total Reward")
plt.show()
```

MONTE CARLO SEARCH TREE SEARCH

```python
import numpy as np
import matplotlib.pyplot as plt
import random

# Step 1: Generate random self-generated data (simulating states, actions, and rewards)
np.random.seed(42)  # for reproducibility
num_data = 1000  # Number of data points
states = np.random.randn(num_data, 1)  # Random states
actions = np.random.randint(0, 2, num_data)  # Random actions (0 or 1)
rewards = np.random.randn(num_data)  # Random rewards (continuous)

# Step 2: Define the MCTS Node and Tree structure
class MCTSNode:
    def __init__(self, state):
        self.state = state  # Current state
        self.parent = None  # Parent node
        self.children = []  # Child nodes
        self.visits = 0  # Number of visits to this node
        self.value = 0  # Cumulative value (reward)
        self.action = None  # Action taken to reach this node

class MCTSTree:
    def __init__(self):
        self.root = MCTSNode(state=None)
        self.visited=set()  # Root node of the tree (initial state)

    def select(self, node):

        # Selection phase: choose a child node to expand based on UCB1

        if len(node.children) == 0:
            return node
        if node in self.visited:  # Access visited using self
            return node
        self.visited.add(node)  # Access visited using self
        ucb_values = [(child.value / (child.visits + 1e-6)) + np.sqrt(2 * np.log(node.visits + 1) / (child.visits + 1e-6)) for child in node.children]
        best_child_index = np.argmax(ucb_values)
        return self.select(node.children[best_child_index])

    def expand(self, node):
        # Expansion phase: add a new child node with a random action
        possible_actions = [0, 1]  # Two possible actions (example: 0 or 1)
        action = random.choice(possible_actions)
        child_state = np.random.randn(1)  # Random new state for simplicity
        child_node = MCTSNode(state=child_state)
        child_node.parent = node
        child_node.action = action
        node.children.append(child_node)

    def simulate(self, node):
        # Simulation phase: return a random reward (simulated environment)
        return random.choice(rewards)

    def backpropagate(self, node, reward):
        # Backpropagation phase: update the value and visits for the node and all its ancestors
        while node is not None:
            node.visits += 1
            node.value += reward
            node = node.parent

    def run(self, num_iterations=100):
        for _ in range(num_iterations):
            node = self.root
            # Selection
            node = self.select(node)
            # Expansion
            self.expand(node)
            # Simulation
            reward = self.simulate(node)
            # Backpropagation
            self.backpropagate(node, reward)

# Step 3: Set up the MCTS algorithm and run it
mcts = MCTSTree()

# Number of MCTS iterations (episodes)
num_episodes = 1000
episode_rewards = []

# Run MCTS for each episode
for episode in range(num_episodes):
    mcts.run(num_iterations=10)  # Run 10 iterations of MCTS in each episode

    # Track the cumulative reward of the root node (root node's value as a proxy for the game score)
    episode_rewards.append(mcts.root.value / (mcts.root.visits + 1e-6))  # Normalize by number of visits

# Step 4: Plot the results (reward per episode)
plt.plot(episode_rewards)
plt.title("Monte Carlo Tree Search Learning Progress")
plt.xlabel("Episode")
plt.ylabel("Average Reward")
plt.show()
```

DQN

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import random
import matplotlib.pyplot as plt
from collections import deque

# Step 1: Generate random self-generated data (simulating states, actions, and rewards)
np.random.seed(42)  # For reproducibility
num_data = 100  # Number of data points
states = np.random.randn(num_data, 1)  # Random states
actions = np.random.randint(0, 2, num_data)  # Random actions (0 or 1)
rewards = np.random.randn(num_data)  # Random rewards (continuous)
next_states = np.random.randn(num_data, 1)  # Random next states

# Step 2: Define the Deep Q-Network (DQN) model
class DQN(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(DQN, self).__init__()
        self.fc1 = nn.Linear(input_dim, 64)  # Fully connected layer
        self.fc2 = nn.Linear(64, output_dim)  # Output layer (Q-values for each action)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        return self.fc2(x)

# Step 3: Define the DQN agent class
class DQNAgent:
    def __init__(self, input_dim, output_dim, epsilon=0.1, gamma=0.99, learning_rate=0.001):
        self.input_dim = input_dim
        self.output_dim = output_dim
        self.epsilon = epsilon  # Exploration-exploitation tradeoff
        self.gamma = gamma  # Discount factor
        self.memory = deque(maxlen=10000)  # Experience replay memory
        self.batch_size = 64  # Batch size for training
        self.learning_rate = learning_rate

        # Initialize the DQN model and optimizer
        self.model = DQN(input_dim, output_dim)
        self.optimizer = optim.Adam(self.model.parameters(), lr=self.learning_rate)
        self.loss_fn = nn.MSELoss()

    def select_action(self, state):
        # Select action based on epsilon-greedy policy
        if random.random() < self.epsilon:
            return random.randint(0, self.output_dim - 1)  # Random action
        else:
            state_tensor = torch.tensor(state, dtype=torch.float32)
            q_values = self.model(state_tensor)
            return torch.argmax(q_values).item()  # Action with highest Q-value

    def store_experience(self, state, action, reward, next_state, done):
        # Store experience in memory
        self.memory.append((state, action, reward, next_state, done))

    def train(self):
        # Train the model using experience replay
        if len(self.memory) < self.batch_size:
            return  # Not enough data to train

        batch = random.sample(self.memory, self.batch_size)
        states, actions, rewards, next_states, dones = zip(*batch)

        states = torch.tensor(states, dtype=torch.float32)
        actions = torch.tensor(actions, dtype=torch.long)
        rewards = torch.tensor(rewards, dtype=torch.float32)
        next_states = torch.tensor(next_states, dtype=torch.float32)
        dones = torch.tensor(dones, dtype=torch.bool)

        # Compute Q-values and target Q-values
        q_values = self.model(states)
        next_q_values = self.model(next_states)
        target_q_values = rewards + self.gamma * torch.max(next_q_values, dim=1)[0] * (~dones)

        # Update Q-values using the Bellman equation
        q_value = q_values.gather(1, actions.view(-1, 1)).squeeze(1)
        loss = self.loss_fn(q_value, target_q_values)

        # Perform a gradient descent step
        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()

# Step 4: Initialize DQN agent and run training
input_dim = 1  # State dimension (for simplicity, using a scalar state)
output_dim = 2  # Action space (0 or 1)
agent = DQNAgent(input_dim, output_dim)

num_episodes = 1000
episode_rewards = []

for episode in range(num_episodes):
    total_reward = 0
    for i in range(num_data):
        state = states[i]
        action = agent.select_action(state)
        reward = rewards[i]
        next_state = next_states[i]
        done = False  # Since it's random, we assume the episode doesn't terminate

        # Store experience in memory
        agent.store_experience(state, action, reward, next_state, done)

        # Train the agent (perform a step of Q-learning)
        agent.train()

        # Update the total reward for this episode
        total_reward += reward

    # Store the total reward for the episode
    episode_rewards.append(total_reward)

# Step 5: Plot the results (reward per episode)
plt.plot(episode_rewards)
plt.title("DQN Learning Progress")
plt.xlabel("Episode")
plt.ylabel("Total Reward")
plt.show()
```

DYNA Q

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_actions = 4  # Number of possible actions (e.g., left, right, up, down)
n_states = 10  # Number of possible states (discretized state space)
episodes = 1000  # Number of episodes for training
alpha = 0.1  # Learning rate for Q-learning
gamma = 0.9  # Discount factor for future rewards
epsilon = 0.1  # Exploration rate for epsilon-greedy policy
n_planning_steps = 5  # Number of planning steps per real experience

# Initialize Q-table and model (state-action transition and reward model)
Q_table = np.zeros((n_states, n_actions))
model = {}

# Generate random reward and state transitions
def get_reward(state, action):
    return np.random.randn()  # Random reward from a normal distribution

# Simulate a random state transition
def get_next_state(state, action):
    return np.random.choice(n_states)  # Random next state

# Choose an action based on epsilon-greedy policy
def choose_action(state):
    if np.random.uniform(0, 1) < epsilon:  # Exploration
        return np.random.choice(n_actions)
    else:  # Exploitation
        return np.argmax(Q_table[state])

# Perform Q-learning update
def q_learning_update(state, action, reward, next_state):
    best_next_action = np.argmax(Q_table[next_state])
    Q_table[state, action] += alpha * (reward + gamma * Q_table[next_state, best_next_action] - Q_table[state, action])

# Perform Dyna-Q planning step (simulate experience using the model)
def dyna_q_planning():
    for _ in range(n_planning_steps):
        # Randomly select a state-action pair from the model
        state, action = list(model.keys())[np.random.choice(len(model))]
        next_state, reward = model[(state, action)]

        # Perform Q-learning update using the simulated experience
        q_learning_update(state, action, reward, next_state)

# Training loop for Dyna-Q
reward_history = []

for episode in range(episodes):
    state = np.random.choice(n_states)  # Random initial state
    total_reward = 0

    # For each episode, run for a set number of steps (maximum 100 steps)
    for step in range(100):
        action = choose_action(state)
        reward = get_reward(state, action)
        total_reward += reward

        # Get next state (random)
        next_state = get_next_state(state, action)

        # Perform Q-learning update with real experience
        q_learning_update(state, action, reward, next_state)

        # Add the experience to the model (state transition and reward)
        model[(state, action)] = (next_state, reward)

        # Perform planning by simulating experience using the model
        dyna_q_planning()

        # Move to the next state
        state = next_state

    reward_history.append(total_reward)

# Plotting the rewards history over episodes
plt.figure(figsize=(10, 6))
plt.plot(reward_history)
plt.xlabel("Episodes")
plt.ylabel("Total Reward")
plt.title("Dyna-Q Reward History over 1000 Episodes")
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import gym
from torch.distributions import Categorical

# Define the Policy Network
class PolicyNetwork(nn.Module):
    def __init__(self, input_dim, action_dim):
        super(PolicyNetwork, self).__init__()
        self.fc1 = nn.Linear(input_dim, 64)
        self.fc2 = nn.Linear(64, 64)
        self.fc3 = nn.Linear(64, action_dim)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        return torch.softmax(self.fc3(x), dim=-1)

# Define the Value Network
class ValueNetwork(nn.Module):
    def __init__(self, input_dim):
        super(ValueNetwork, self).__init__()
        self.fc1 = nn.Linear(input_dim, 64)
        self.fc2 = nn.Linear(64, 64)
        self.fc3 = nn.Linear(64, 1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        return self.fc3(x)

# PPO Agent Class
class PPOAgent:
    def __init__(self, input_dim, action_dim, lr=3e-4, gamma=0.99, epsilon=0.2, entropy_weight=0.01):
        self.policy_net = PolicyNetwork(input_dim, action_dim)
        self.value_net = ValueNetwork(input_dim)
        self.optimizer_policy = optim.Adam(self.policy_net.parameters(), lr=lr)
        self.optimizer_value = optim.Adam(self.value_net.parameters(), lr=lr)
        self.gamma = gamma
        self.epsilon = epsilon
        self.entropy_weight = entropy_weight

    def select_action(self, state):
        state = torch.tensor(state, dtype=torch.float32)
        probs = self.policy_net(state)
        dist = Categorical(probs)
        action = dist.sample()
        return action.item(), dist.log_prob(action), dist.entropy()






    def compute_advantages(self, rewards, values, next_values, dones):
        # Convert lists to NumPy arrays for element-wise operations
        values = np.array([v.detach().numpy() for v in values], dtype=np.float32)
        next_values = np.array([v.detach().numpy() for v in next_values], dtype=np.float32)
        dones = np.array(dones, dtype=np.float32)

        deltas = rewards + self.gamma * next_values * (1 - dones) - values
        advantages = []
        adv = 0
        for delta in reversed(deltas):
            adv = delta + self.gamma * adv
            advantages.insert(0, adv)
        return torch.tensor(advantages, dtype=torch.float32)

    def update(self, states, actions, old_log_probs, advantages, returns):
        # Compute the new log probabilities
        states = torch.tensor(states, dtype=torch.float32)
        actions = torch.tensor(actions, dtype=torch.long)
        old_log_probs = torch.tensor(old_log_probs, dtype=torch.float32)

        # Forward pass
        new_probs = self.policy_net(states)
        dist = Categorical(new_probs)
        new_log_probs = dist.log_prob(actions)
        entropy = dist.entropy().mean()

        # Compute the ratio (probability ratio between new and old policy)
        ratio = torch.exp(new_log_probs - old_log_probs)

        # Compute the clipped objective
        surrogate_loss = torch.min(ratio * advantages,
                                   torch.clamp(ratio, 1 - self.epsilon, 1 + self.epsilon) * advantages)

        # Policy loss
        policy_loss = -surrogate_loss.mean() - self.entropy_weight * entropy

        # Value loss (using MSE)
        values = self.value_net(states)
        value_loss = nn.MSELoss()(values.squeeze(), returns)

        # Total loss
        loss = policy_loss + value_loss

        # Backpropagate and update
        self.optimizer_policy.zero_grad()
        self.optimizer_value.zero_grad()
        loss.backward()
        self.optimizer_policy.step()
        self.optimizer_value.step()

# Training Loop
def train(env, agent, episodes=1000, max_timesteps=500, batch_size=32):
    for episode in range(episodes):
        states, actions, log_probs, rewards, dones = [], [], [], [], []
        values, next_values = [], []

        state = env.reset()
        for t in range(max_timesteps):
            action, log_prob, entropy = agent.select_action(state)
            value = agent.value_net(torch.tensor(state, dtype=torch.float32)).squeeze()

            next_state, reward, done, _ = env.step(action)

            states.append(state)
            actions.append(action)
            log_probs.append(log_prob)
            rewards.append(reward)
            dones.append(done)
            values.append(value)

            state = next_state

            if done:
                break

        next_value = agent.value_net(torch.tensor(next_state, dtype=torch.float32)).squeeze()
        next_values = values[1:] + [next_value]

        # Compute returns (sum of rewards)
        returns = []
        R = next_value
        for r, d in zip(reversed(rewards), reversed(dones)):
            R = r + agent.gamma * R * (1 - d)
            returns.insert(0, R)
        returns = torch.tensor(returns, dtype=torch.float32)

        # Compute advantages (using GAE)
        advantages = agent.compute_advantages(rewards, values, next_values, dones)

        # Update the agent
        agent.update(states, actions, log_probs, advantages, returns)
        print(f'Episode {episode+1}/{episodes} complete')

# Example usage
if __name__ == "__main__":
    env = gym.make("CartPole-v1")  # You can replace with any other environment
    agent = PPOAgent(input_dim=env.observation_space.shape[0], action_dim=env.action_space.n)
    train(env, agent)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from torch.distributions import Categorical

# Generate synthetic data (1000 samples)
def generate_synthetic_data(num_samples=1000):
    states = np.random.rand(num_samples, 1)  # 1000 random states
    actions = np.random.choice([0, 1], num_samples)  # Random actions (0 or 1)
    rewards = np.random.rand(num_samples)  # Random rewards
    next_states = np.random.rand(num_samples, 1)  # Next state
    dones = np.random.choice([0, 1], num_samples)  # Random terminal states (0 or 1)
    return states, actions, rewards, next_states, dones

# Define the Policy Network
class PolicyNetwork(nn.Module):
    def __init__(self, input_dim, action_dim):
        super(PolicyNetwork, self).__init__()
        self.fc1 = nn.Linear(input_dim, 64)
        self.fc2 = nn.Linear(64, 64)
        self.fc3 = nn.Linear(64, action_dim)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        return torch.softmax(self.fc3(x), dim=-1)

# Define the Value Network
class ValueNetwork(nn.Module):
    def __init__(self, input_dim):
        super(ValueNetwork, self).__init__()
        self.fc1 = nn.Linear(input_dim, 64)
        self.fc2 = nn.Linear(64, 64)
        self.fc3 = nn.Linear(64, 1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        return self.fc3(x)

# PPO Agent Class
class PPOAgent:
    def __init__(self, input_dim, action_dim, lr=3e-4, gamma=0.99, epsilon=0.2, entropy_weight=0.01):
        self.policy_net = PolicyNetwork(input_dim, action_dim)
        self.value_net = ValueNetwork(input_dim)
        self.optimizer_policy = optim.Adam(self.policy_net.parameters(), lr=lr)
        self.optimizer_value = optim.Adam(self.value_net.parameters(), lr=lr)
        self.gamma = gamma
        self.epsilon = epsilon
        self.entropy_weight = entropy_weight

    def select_action(self, state):
        state = torch.tensor(state, dtype=torch.float32)
        probs = self.policy_net(state)
        dist = Categorical(probs)
        action = dist.sample()
        return action.item(), dist.log_prob(action), dist.entropy()

    def compute_advantages(self, rewards, values, next_values, dones):
    # Convert lists to NumPy arrays for element-wise operations
        rewards = np.array(rewards, dtype=np.float32)
        values = np.array([v.detach().numpy() for v in values], dtype=np.float32) # Detach values before converting to NumPy
        next_values = np.array([v.detach().numpy() for v in next_values], dtype=np.float32) # Detach next_values before converting to NumPy
        dones = np.array(dones, dtype=np.float32)

        deltas = rewards + self.gamma * next_values * (1 - dones) - values
        advantages = []
        adv = 0
        for delta in reversed(deltas):
            adv = delta + self.gamma * adv
            advantages.insert(0, adv)
        return torch.tensor(advantages, dtype=torch.float32)

    def update(self, states, actions, old_log_probs, advantages, returns):
        # Compute the new log probabilities
        states = torch.tensor(states, dtype=torch.float32)
        actions = torch.tensor(actions, dtype=torch.long)
        old_log_probs = torch.tensor(old_log_probs, dtype=torch.float32)

        # Forward pass
        new_probs = self.policy_net(states)
        dist = Categorical(new_probs)
        new_log_probs = dist.log_prob(actions)
        entropy = dist.entropy().mean()

        # Compute the ratio (probability ratio between new and old policy)
        ratio = torch.exp(new_log_probs - old_log_probs)

        # Compute the clipped objective
        surrogate_loss = torch.min(ratio * advantages,
                                   torch.clamp(ratio, 1 - self.epsilon, 1 + self.epsilon) * advantages)

        # Policy loss
        policy_loss = -surrogate_loss.mean() - self.entropy_weight * entropy

        # Value loss (using MSE)
        values = self.value_net(states)
        value_loss = nn.MSELoss()(values.squeeze(), returns)

        # Total loss
        loss = policy_loss + value_loss

        # Backpropagate and update
        self.optimizer_policy.zero_grad()
        self.optimizer_value.zero_grad()
        loss.backward()
        self.optimizer_policy.step()
        self.optimizer_value.step()

# Training Loop
def train(agent, states, actions, rewards, next_states, dones, episodes=1000, batch_size=32):
    episode_rewards = []
    for episode in range(episodes):
        states_batch, actions_batch, log_probs_batch, rewards_batch, dones_batch = [], [], [], [], []
        values_batch, next_values_batch = [], []

        for i in range(len(states)):
            state = states[i]
            action, log_prob, entropy = agent.select_action(state)
            value = agent.value_net(torch.tensor(state, dtype=torch.float32)).squeeze()

            next_state = next_states[i]
            reward = rewards[i]
            done = dones[i]

            states_batch.append(state)
            actions_batch.append(action)
            log_probs_batch.append(log_prob)
            rewards_batch.append(reward)
            dones_batch.append(done)
            values_batch.append(value)

            if done:
                break

        next_value = agent.value_net(torch.tensor(next_state, dtype=torch.float32)).squeeze()
        next_values_batch = values_batch[1:] + [next_value]

        # Compute returns (sum of rewards)
        returns = []
        R = next_value
        for r, d in zip(reversed(rewards_batch), reversed(dones_batch)):
            R = r + agent.gamma * R * (1 - d)
            returns.insert(0, R)
        returns = torch.tensor(returns, dtype=torch.float32)

        # Compute advantages (using GAE)
        advantages = agent.compute_advantages(rewards_batch, values_batch, next_values_batch, dones_batch)

        # Update the agent
        agent.update(states_batch, actions_batch, log_probs_batch, advantages, returns)

        episode_rewards.append(sum(rewards_batch))  # Track the cumulative reward for this episode
        print(f'Episode {episode+1}/{episodes} complete, Cumulative Reward: {sum(rewards_batch)}')

    return episode_rewards

# Generate synthetic data
states, actions, rewards, next_states, dones = generate_synthetic_data()

# Initialize PPO agent
agent = PPOAgent(input_dim=1, action_dim=2)

# Train the agent and get the rewards per episode
episode_rewards = train(agent, states, actions, rewards, next_states, dones, episodes=50)

# Plot the rewards
plt.plot(episode_rewards)
plt.xlabel('Episode')
plt.ylabel('Cumulative Reward')
plt.title('PPO Training - Cumulative Rewards per Episode')
plt.show()
```
