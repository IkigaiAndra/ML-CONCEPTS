```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Generate 2000 data samples with 5 features
np.random.seed(42)  # Set seed for reproducibility
data = np.random.rand(2000, 5)

# Create a Pandas DataFrame
df = pd.DataFrame(data, columns=['Feature1', 'Feature2', 'Feature3', 'Feature4', 'Feature5'])

# Standardize the features to have zero mean and unit variance
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Apply PCA with 5 components
pca = PCA(n_components=5)
principal_components = pca.fit_transform(scaled_data)

# Create a new DataFrame with the principal components
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5'])

# Plot the first two principal components
plt.figure(figsize=(8, 6))
plt.scatter(principal_df['PC1'], principal_df['PC2'], alpha=0.5)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA with 5 Components (First Two PCs)')
plt.grid(True)
plt.show()

# Plot the explained variance ratio
plt.figure(figsize=(8, 6))
plt.plot(np.arange(1, pca.n_components_ + 1), pca.explained_variance_ratio_, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratio vs. Number of Components')
plt.grid(True)
plt.show()
```
