```python
!pip install mlxtend==0.20.0  # Install necessary library
!pip install matplotlib==3.7.1 # Install matplotlib for plotting

import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import matplotlib.pyplot as plt
import random

# Generate sample data
itemset = ['apple', 'banana', 'orange', 'milk', 'bread', 'eggs']
transactions = []
for _ in range(100):
    num_items = random.randint(1, 5)
    transaction = random.sample(itemset, num_items)
    transactions.append(transaction)

# Transform data into the required format
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

# Apply Apriori algorithm
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Filter rules for plotting
plot_rules = rules[(rules['lift'] >= 1.2) & (rules['confidence'] >= 0.5)]

# Plot the rules
plt.figure(figsize=(10, 6))
plt.scatter(plot_rules['support'], plot_rules['confidence'], alpha=0.5)
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Association Rules')

# Annotate the points with rule information
for index, row in plot_rules.iterrows():
    plt.annotate(f"{list(row['antecedents'])} -> {list(row['consequents'])}",
                 (row['support'], row['confidence']),
                 fontsize=8, ha='center')

plt.show()
```

*quantitative association rule minning*

```python
!pip install mlxtend==0.20.0  # Install necessary library
!pip install matplotlib==3.7.1 # Install matplotlib for plotting

import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import matplotlib.pyplot as plt
import random
import numpy as np

# Generate sample data with quantitative attributes
num_transactions = 100
itemset = ['apple', 'banana', 'orange', 'milk', 'bread', 'eggs']
transactions = []
quantities = {}  # Store quantities for each item in each transaction

for _ in range(num_transactions):
    num_items = random.randint(1, 5)
    transaction = random.sample(itemset, num_items)
    transactions.append(transaction)
    for item in itemset:
        quantities.setdefault(item, []).append(0)
    for item in transaction:
        quantities[item][-1]=random.randint(1, 5)



# Create a DataFrame for quantitative data
df_quant = pd.DataFrame(quantities)
df_quant['transaction_id'] = range(num_transactions)

# Discretize quantitative data into bins
for item in itemset:
    # Determine the number of unique bins after dropping duplicates
    num_bins = len(pd.qcut(df_quant[item], q=3, duplicates='drop').cat.categories)

    # Generate labels based on the number of unique bins
    labels = [f'bin_{i}' for i in range(1, num_bins + 1)]  # or any other label scheme you prefer

    # Apply qcut with dynamically generated labels
    df_quant[item + '_bin'] = pd.qcut(df_quant[item], q=3, labels=labels, duplicates='drop')

# Prepare data for association rule mining
transactions_binned = []
for i in range(num_transactions):
    transaction_binned = []
    for item in itemset:
        if df_quant.loc[i, item + '_bin'] != np.nan:
            transaction_binned.append(item + '=' + str(df_quant.loc[i, item + '_bin']))
    transactions_binned.append(transaction_binned)

te = TransactionEncoder()
te_ary = te.fit(transactions_binned).transform(transactions_binned)
df_binned = pd.DataFrame(te_ary, columns=te.columns_)

# Apply Apriori algorithm
frequent_itemsets = apriori(df_binned, min_support=0.2, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Filter rules for plotting
plot_rules = rules[(rules['lift'] >= 1.2) & (rules['confidence'] >= 0.5)]

# Plot the rules
plt.figure(figsize=(10, 6))
plt.scatter(plot_rules['support'], plot_rules['confidence'], alpha=0.5)
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Quantitative Association Rules')

# Annotate the points with rule information
for index, row in plot_rules.iterrows():
    plt.annotate(f"{list(row['antecedents'])} -> {list(row['consequents'])}",
                 (row['support'], row['confidence']),
                 fontsize=8, ha='center')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import random

# Step 1: Generate synthetic transaction data
itemset = ['apple', 'banana', 'orange', 'milk', 'bread', 'eggs']
transactions = []
for _ in range(100):
    num_items = random.randint(1, 5)
    transaction = random.sample(itemset, num_items)
    transactions.append(transaction)

# Step 2: Build the vertical data representation (TID-Itemset matrix)
item_tid_map = defaultdict(set)

# Populate the map with transaction IDs (TID) for each item
for tid, transaction in enumerate(transactions):
    for item in transaction:
        item_tid_map[item].add(tid)

# Step 3: ECLAT algorithm to find frequent itemsets
def eclat(item_tid_map, min_support):
    frequent_itemsets = []

    def depth_first_search(prefix, item_tid_map_subset):
        nonlocal frequent_itemsets

        if len(prefix) > 0 and len(item_tid_map_subset) >= min_support:  # Changed condition
            frequent_itemsets.append((prefix, len(item_tid_map_subset)))

        for item, tid_set in item_tid_map.items():
            if item not in prefix:  # Avoid redundant itemsets
                common_tids = item_tid_map_subset.intersection(tid_set)

                if len(common_tids) >= min_support:
                    new_prefix = prefix + [item]
                    depth_first_search(new_prefix, common_tids)

    depth_first_search([], set(range(len(transactions))))
    return frequent_itemsets

# Step 4: Generate and print frequent itemsets
min_support = 20  # Define the minimum support threshold (e.g., 20%)
frequent_itemsets = eclat(item_tid_map, min_support)

# Print the frequent itemsets and their supports
print("Frequent Itemsets (with support >= {}):".format(min_support))
for itemset, support in frequent_itemsets:
    print(f"{itemset}: Support = {support}")

# Step 5: Visualize the frequent itemsets' support in a plot
itemsets_str = [str(itemset) for itemset, _ in frequent_itemsets]
supports = [support for _, support in frequent_itemsets]

plt.figure(figsize=(10, 6))
plt.barh(itemsets_str, supports, color='skyblue')
plt.xlabel('Support')
plt.ylabel('Itemsets')
plt.title('Frequent Itemsets Support (ECLAT Algorithm)')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

# Step 1: Generate a synthetic dataset with 50,000 transactions
# Let's simulate a dataset with 100 items, some of which will be rare
num_transactions = 50000
num_items = 100

# Randomly generating transactions with 50 items (out of 100 possible items)
transactions = []
for _ in range(num_transactions):
    # Randomly decide how many items are in the transaction (between 1 and 10 items)
    num_items_in_transaction = np.random.randint(1, 11)
    transaction = np.random.choice([f'Item_{i}' for i in range(1, num_items + 1)], num_items_in_transaction, replace=False)
    transactions.append(list(transaction))

# Step 2: Convert the data into a DataFrame using TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

# Step 3: Apply the Apriori algorithm to find frequent itemsets
# Use a low minimum support to find rare itemsets (e.g., 0.01 or 1%)
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)

# Step 4: Generate association rules with a low support threshold but high confidence
# We will use the 'lift' metric and set the minimum lift to 1
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Filter the rules to focus on rare items by ensuring low support and high confidence
rare_rules = rules[(rules['support'] < 0.05) & (rules['confidence'] > 0.6)]

# Step 5: Display the rare association rules
print("Rare Association Rules (Support < 5% and Confidence > 60%):")
print(rare_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])

# Step 6: Visualize the rare association rules (Support vs Confidence)
plt.figure(figsize=(10, 6))
plt.scatter(rare_rules['support'], rare_rules['confidence'], c=rare_rules['lift'], cmap='viridis', s=50)
plt.colorbar(label='Lift')
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Rare Association Rules: Support vs Confidence')
plt.show()
```
