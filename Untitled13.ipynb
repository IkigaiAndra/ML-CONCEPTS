 np.meshgrid() and then plotted using p1.contour()

```python
import numpy as np
import matplotlib.pyplot as plt

# Create the meshgrid
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(x, y)

# Calculate the values for the contour plot
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create the contour plot
fig, ax = plt.subplots()  # Added this line
p1 = ax.contour(X, Y, Z)  # Changed this line to use p1 = ...
# ax.clabel(p1, inline=1, fontsize=10)  # Optional: add contour labels
#fig.colorbar(p1, ax=ax)  # Optional: add a colorbar
#ax.set_title('Contour Plot')  # Optional: add a title
#ax.set_xlabel('X')  # Optional: add an x-axis label
#ax.set_ylabel('Y')  # Optional: add a y-axis label

# Show the plot
plt.show()  # Or fig.show() if you prefer
```
