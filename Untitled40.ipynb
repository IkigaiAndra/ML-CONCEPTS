```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.svm import OneClassSVM
from sklearn.semi_supervised import SelfTrainingClassifier

# Generate data
X, y = make_blobs(n_samples=300, centers=2, cluster_std=0.8, random_state=42)
X_train = np.concatenate([X[y == 0], X[y == 1][:5]])  # Normal data + a few anomalies
X_test = np.concatenate([X[y == 1][5:], X[y == 0][:10]]) # Anomalies + a few normal data
y_train = np.zeros(X_train.shape[0]) # Labels for training (0 for normal)
y_test = np.ones(X_test.shape[0]) # Labels for testing (1 for anomaly)
# Initialize base classifier and self-training classifier
base_clf = OneClassSVM(nu=0.1)
clf = SelfTrainingClassifier(base_clf)

# Train the model
clf.fit(X_train, y_train)

# Predict anomalies on test data
y_pred = clf.predict(X_test)
```

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.svm import OneClassSVM

# Generate data
X, y = make_blobs(n_samples=300, centers=2, cluster_std=0.8, random_state=42)
X_train = np.concatenate([X[y == 0], X[y == 1][:5]])  # Normal data + a few anomalies
X_test = np.concatenate([X[y == 1][5:], X[y == 0][:10]]) # Anomalies + a few normal data
y_train = np.zeros(X_train.shape[0]) # Labels for training (0 for normal)
y_test = np.ones(X_test.shape[0]) # Labels for testing (1 for anomaly)

# Initialize and train OneClassSVM directly
clf = OneClassSVM(nu=0.1)
clf.fit(X_train)

# Predict anomalies on test data
y_pred = clf.predict(X_test)

# Adjust predictions (OneClassSVM returns -1 for anomalies)
y_pred[y_pred == 1] = 0  # Normal data
y_pred[y_pred == -1] = 1 # Anomaly data
# Plot the results
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred, cmap='viridis')
plt.title('Self-Training with One-Class Classifier')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Anomaly Prediction (0: Normal, 1: Anomaly)')
plt.show()
```
